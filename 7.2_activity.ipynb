Scatterplot

```python
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds = diamonds[
            diamonds.cut.isin(['Premium', 'Good']) &
            diamonds.color.isin(['D','F','J'])
            ].sample(n=100, random_state=22)

```

```python
diamonds.shape
```

```python
diamonds.head()
```

Basics

```python
sns.set_style('dark')
```

```python
sns.scatterplot(diamonds.carat, diamonds.price)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds)
```

```python
sns.scatterplot(x='x', y='y', data=diamonds)
```

Hue

```python
sns.scatterplot(x='carat', 
                y='price',
                hue = 'cut',
                data=diamonds
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                hue = 'cut',
                palette = ['purple', '#55CCCC'],
                data=diamonds
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                hue = 'depth',
                data=diamonds
                 )
```

Style

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                s=100
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                style = 'cut',
                s=100
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                size = 'cut',
                s=100
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                size = 'cut',
                sizes=[150,50],
                s=100
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                size = 'depth',
                s=100
                 )
```

hue + stlye + size

```python
sns.scatterplot(x='carat', 
                y='price',
                hue = 'cut',
                style = 'cut',
                size = 'cut',
                data=diamonds,
                s=100
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                hue = 'cut',
                style = 'color',
                size = 'cut',
                data=diamonds,
                s=100
                 )
```

Additional styling

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                s=100,
                alpha=0.6
                 )
```

Marker stlying

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                s=200,
                marker='*'
                 )
```

```python
sns.scatterplot(x='carat', 
                y='price',
                data=diamonds,
                s=200,
                marker='*',
                edgecolor='black'
                 )
```

Strippplot

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.stripplot(cars.weight)
```

```python
sns.stripplot(cars.weight, cars.origin)
```

```python
sns.stripplot(x = 'weight',
              y = 'origin',
              data = cars
              )
```

```python
sns.stripplot(x = 'weight',
              y = 'origin',
              hue = 'cylinders',
              data = cars
              )
```

```python
sns.stripplot(x = 'weight',
              y = 'origin',
              hue = 'cylinders',
              dodge = True, #split up hues onto their own plots
              data = cars
              )
```

```python
sns.stripplot(x = 'origin',
              y = 'weight',
              hue = 'cylinders',
              dodge = True, #split up hues onto their own plots
              data = cars
              )
```

Styling

Jitter

```python
sns.stripplot(cars.weight,
              cars.origin,
              jitter=False
              )
```

```python
sns.stripplot(cars.weight,
              cars.origin,
              jitter=0.05
              )
```

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
    sns.stripplot(cars.weight, 
                  cars.origin,
                  jitter = jitter
                 )
                  
```

```python
interactive(jitter_widget, jitter = (0, 0.65, 0.05))
```

Alpha

```python
sns.set_style('whitegrid')
```

```python
sns.stripplot(cars.weight,
              cars.origin, 
              alpha = 0.5
              )
```

```python
sns.stripplot(cars.weight,
              cars.origin, 
              alpha = 0.3,
              size = 8
              )
```

linewidth, edgecolor

```python
sns.stripplot(cars.weight,
              cars.origin, 
              linewidth = 1
              )
```

```python
sns.stripplot(cars.weight,
              cars.origin, 
              linewidth = 1,
              edgecolor = 'black'
              )
```

Marker, same options as scatterplot

```python
sns.stripplot(cars.weight,
              cars.origin, 
              alpha = 0.7,
              size = 10,
              jitter = 0.25,
              marker = '*'
              )
```
